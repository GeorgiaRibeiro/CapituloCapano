## Criando função de pré-processamento do texto e cálculo da frequencia relativa

```python
def preprocess_and_find_top_words_with_frequencies(text, top_n=10):
    # Preprocess the text: remove punctuation, lowercase, and tokenize
    text = re.sub(r'[^\w\s]', '', text.lower())  # Remove punctuation and lowercase
    words = word_tokenize(text)  # Tokenize the text

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]

    # Count word frequencies
    word_freq = Counter(filtered_words)

    # Calculate total words to get relative frequencies
    total_words = sum(word_freq.values())

    # Get the top N most common words and their relative frequencies
    most_common_words = word_freq.most_common(top_n)
    relative_frequencies = [(word, freq / total_words) for word, freq in most_common_words]

    return relative_frequencies
```

## Aplicando a função no df

```python
df2 = df
```

```python
df2['top_words_with_freq'] = df2['abstract'].apply(preprocess_and_find_top_words_with_frequencies)

new_rows = []
for _, row in df2.iterrows():
    for word, freq in row['top_words_with_freq']:
        new_rows.append({'id': row['id'], 'word': word, 'relative_frequency': freq})

# Create the new DataFrame
freq_df = pd.DataFrame(new_rows)

print(freq_df)
```

## Top 20 palavras por tema

### calculo

```python
# resgatar coluna do tema
df_class_words = pd.merge(df2, freq_df, on='id', how='left')\
                   .loc[:, ['id', 'classificacao_llm_final', 'word', 'relative_frequency']]

# somar frequencia de palavras por tema
df_class_words2 = df_class_words.groupby(['classificacao_llm_final','word'])['relative_frequency'].sum().reset_index(name='relative_frequency_bytheme')
```

```python
df_class_words.head()
```

```python
df_class_words2[df_class_words2['classificacao_llm_final'] == 'Climate Change'].sort_values(['relative_frequency_bytheme'], ascending = False)
```

```python
# conferir qtd de palavras por tema
df_class_words.groupby(['classificacao_llm_final']).size().reset_index(name='count_words').sort_values(['count_words'], ascending=False)
```

### selecionar top 20 mais frequentes

```python
# ordernar frequencia por grupo
grouped = df_class_words2.sort_values(by=['classificacao_llm_final', 'relative_frequency_bytheme'], ascending=[True, False])

# ranquear as palavras por grupo
grouped['rank'] = grouped.groupby('classificacao_llm_final')['relative_frequency_bytheme'].rank(method='first', ascending=False)
```

```python
grouped.head()
```

```python
# Filter the top '20' rows in each group and drop auxiliar column
df_class_words2 = grouped[grouped['rank'] <= 20].drop(columns=['rank'])
```

```python
themes = df_class_words2['classificacao_llm_final'].unique().tolist()

print('Top 20 - Palavras mais frequentes por tema: \n')
for theme in themes:
  word_list = df_class_words2[df_class_words2['classificacao_llm_final'] == theme]['word'].values.tolist()
  print(theme)
  for i in range(0, len(word_list), 10):   # break line for list after the item n. 10
    print(', '.join(word_list[i: i + 10]))
  print('\n')
```
